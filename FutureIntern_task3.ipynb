```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df= pd.read_csv("/content/sample_data/Iris.csv")
```

```python
print(iris_df.head())
```

```python
sns.set(style="whitegrid")
```

```python
# 1. Histogram for Sepal Length
plt.figure(figsize=(10, 6))
sns.histplot(iris_df['SepalLengthCm'], bins=20, kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Histogram for Sepal Width
plt.figure(figsize=(10, 6))
sns.histplot(iris_df['SepalWidthCm'], bins=20, kde=True)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Histogram for Petal Length
plt.figure(figsize=(10, 6))
sns.histplot(iris_df['PetalLengthCm'], bins=20, kde=True)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Histogram for Petal Width
plt.figure(figsize=(10, 6))
sns.histplot(iris_df['PetalWidthCm'], bins=20, kde=True)
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Bar chart for Species Count
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=iris_df)
plt.title("Species Distribution in the Iris Dataset")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
